```python
delta = 1/8/32
print(f"delta={delta}")
delta_i = -delta
delta_q = delta
k7 = np.array([1, -9, +45, 0, -45, +9, -1]) / 60.0

xi = xi_osr8 * 1/(1+delta)
xq = xq_osr8 * 1/(1+delta)
```

```python
def frac_delay_1st(x, delta, kernel):
    """
    Single-term Taylor fractional delay: y = f + f', where f' is conv(x, kernel).
    kernel must sum to zero and approximate the derivative.
    """
    f0 = x
    f1 = np.convolve(x, kernel, mode='same')
    return f0 + delta*f1

xi_ = signal.lfilter(k7, 1.0, xi)
xq_ = signal.lfilter(k7, 1.0, xq)

xi_delay = xi + delta_i*xi_
xq_delay = xq + delta_q*xq_

np.savetxt("../../../../VHDL/data/delay_test/xin_i_delay_2.txt", xi_delay, fmt="%.12f")
np.savetxt("../../../../VHDL/data/delay_test/xin_q_delay_2.txt", xq_delay, fmt="%.12f")

xi_delay_1 = frac_delay_1st(xi, delta_i, k7)
xq_delay_1 = frac_delay_1st(xq, delta_q, k7)
```
