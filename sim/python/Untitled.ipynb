```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import freqz

# Define filter coefficients
b = np.array([3.19382974, -8.02002256, 8.73762976, -4.61756997, 0.97458298])
a = np.array([1, -1.63632004, 1.47600867, -0.75840147, 0.2125798, -0.02541702])

a /= 10
b /= 10
a[0] = 1

# Compute the frequency response
w, h = freqz(b, a, worN=8000)

# Plot the amplitude response
plt.figure()
plt.plot(w / np.pi, 20 * np.log10(abs(h)))  # Convert magnitude to dB
plt.title('IIR Filter Frequency Response')
plt.xlabel('Normalized Frequency (×π rad/sample)')
plt.ylabel('Amplitude [dB]')
plt.grid()
plt.show()
```
