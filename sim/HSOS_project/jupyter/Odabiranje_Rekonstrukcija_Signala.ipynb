# <center> Odabiranje signala povorkom Dirakovih impulsa
    
Teorema o odabiranju se može pokazati preko odabiranja povorkom Dirakovih impulsa (Dirakov češalj, engl. Dirac Comb)
$$
\mr{III}(t) =  \sum\limits_{n=-\infty}^{\infty}  \delta(t-nT)
$$

$$
x(t) = x_\mr{a}(t) \sum\limits_{n=-\infty}^{\infty}  \delta(t-nT)
$$

Signal $x(t)$ je povorka Dirakovih impulsa $x_\mr{a}(nT)\delta(t-nT)$

<img src="images/sampling_dirac.svg" />


Spektar signala $X(f)$ se dobija Furijeovom transformacijom
\begin{align*}
X(f)
&= 
\mathcal{F} \left\lbrace x_\mr{a}(t) \sum\limits_{n=-\infty}^{\infty}  \delta(t-nT) \right\rbrace (f)\\
&=
X_\mr{a}(f) * \mathcal{F} \left\lbrace  \sum\limits_{n=-\infty}^{\infty}  \delta(t-nT) \right\rbrace (f) \\
&= \int\limits_{-\infty}^{+\infty} X_\mr{a}(f-u) \mathcal{F} \left\lbrace  \sum\limits_{n=-\infty}^{\infty}  \delta(t-nT) \right\rbrace (u) \dd u
\end{align*}
Upotrebljeno je svojstvo Furijeove transformacije
$$
\mathcal{F} \left\lbrace g(t)h(t)  \right\rbrace (f) = \mathcal{F} \left\lbrace g(t)\right\rbrace (f) * \mathcal{F} \left\lbrace h(t)  \right\rbrace (f)
$$

Povorka Dirakovih impulsa
\begin{align*}
\mr{III}(t) =  \sum\limits_{n=-\infty}^{\infty}  \delta(t-nT) 
\end{align*}
je periodična funkcija i može se razviti u Furijeov red
$$
\mr{III}(t) =  \sum\limits_{n=-\infty}^{\infty} c_k e^{\jj 2 \pi f_\mr{s} n t }
$$
$$
c_n = \frac{1}{T}\int\limits_{-\frac{T}{2}}^{\frac{T}{2}} \delta(t) e^{\jj 2\pi f_\mr{s} n t} = \frac{1}{T}
$$
$$
\mr{III}(t) =  \frac{1}{T} \sum\limits_{n=-\infty}^{\infty}  e^{\jj 2 \pi f_\mr{s} n t }
$$

Furijeova transformacija $\mr{III}(f)$
\begin{align*}
\mr{III}(f) &= \mathcal{F}\left\lbrace \frac{1}{T} \sum\limits_{n=-\infty}^{\infty}  e^{\jj 2 \pi f_\mr{s} n t }  \right\rbrace(f) \\
&= \frac{1}{T} \sum\limits_{n=-\infty}^{\infty} \mathcal{F}\left\lbrace e^{\jj 2 \pi f_\mr{s} n t }  \right\rbrace(f) \\
&= \frac{1}{T} \sum\limits_{n=-\infty}^{\infty} \delta(f-n f_\mr{s})
\end{align*}
U drugom koraku upotrebljeni su transformacioni parovi
$$
1 \leftrightarrow \delta(f)
$$
$$
x(t) e^{\jj 2 \pi f_0 t  } \leftrightarrow X(f-f_0)
$$

\begin{align*}
X(f)
&= 
\frac{1}{T}
\int\limits_{-\infty}^{+\infty} X_\mr{a}(f-u) 
\sum\limits_{n=-\infty}^{\infty} \delta(u-n f_\mr{s}) \dd u \\
&= 
\frac{1}{T}
\sum\limits_{n=-\infty}^{\infty}
\int\limits_{-\infty}^{+\infty} X_\mr{a}(f-u) \delta(u-n f_\mr{s}) \dd u \\
&=
\frac{1}{T}
\sum\limits_{n=-\infty}^{\infty} X_\mr{a}(f-n f_\mr{s})
\end{align*}

U poslednjem koraku upotrebljeno je svojstvo Dirakove funkcije
$$
\int\limits_{-\infty}^{+\infty} f(x) \delta(x-x_0) \dd x = f(x_0)
$$

## <center> Odabiranje signala u višim Nikvistovim zonama

Teorema odabiranja je izvedena pod pretpostavkom da signal zauzima opseg učestanosti
$(-B,B)$ kada je potrebno da učestanost odabiranja zadovolji uslov $f_\mr{s} \geq 2B$.



Ukoliko je spektar signala ograničen na opseg $(f_\mr{L}, f_\mr{H})$ moguće je izvršiti odabiranje signala sa $f_\mr{s} < f_\mr{H}$.


Odabiranje u višim Nikvistovim zonama je ograničeno propusnim opsegom A/D konvertora (enlg. full analog bandwidth) i podrhtavanjem takta odabiranja (engl. jitter).

<img src="images/adc_sampling_bandpass.svg" />

Pretpostavimo da je spektar signala ograničen na opseg
$$
B = f_\mr{H} - f_\mr{L}
$$
Odnos maksimalne učestanosti i opsega $m$ je
$$
m = \frac{f_\mr{H}}{B}
$$
Ukoliko je $m$ ceo broj, moguće je izvršiti odabiranje učestanošću
$$
f_\mr{s} = 2B
$$

<img src="images/adc_nyquist_integer.svg" />

U opštem slučaju $f_\mr{s}$ se bira tako da se $(k-1)$-va i $k$-ta replika signala ne preklapaju sa željenim signalom.

Do preklapanja signala ne dolazi ukoliko su zadovoljeni uslovi:
\begin{align*}
2 f_\mr{H} &\leq k f_\mr{s} \\
(k-1) f_\mr{s} &\leq 2 f_\mr{L}
\end{align*}
$$
\frac{2 f_\mr{H}}{k} \leq f_\mr{s}   \leq \frac{2 f_\mr{L}}{k-1}
$$
<img src="images/adc_nyquist_frac.svg" />


Prva nejednačina se može preurediti u
$$
\frac{1}{f_\mr{s}} \leq  \frac{k}{2 f_\mr{H}}
$$
Pošto je $f_\mr{L} = f_\mr{H}-B$ dobijamo
$$
(k-1)f_\mr{s} \leq 2 f_\mr{H} - 2B
$$
Množenjem levih i desnih strana nejednačine i rešavanjem po $k$
$$
k_\mr{max} \leq \frac{f_\mr{H}}{B}
$$
$$
k_\mr{max} = \left\lfloor \frac{f_\mr{H}}{B} \right\rfloor
$$

Izbor Nikvistove zone je ograničen sa
$$
1 \leq k \leq \left\lfloor \frac{f_\mr{H}}{B} \right\rfloor
$$
Učestanost odabiranja je ograničena sa
$$
\frac{2 f_\mr{H}}{k} \leq f_\mr{s} \leq \frac{2 f_\mr{L}}{k-1}
$$
Minimalna učestanost odabiranja u $k$-toj Nikvistovoj zoni je
$$
f_\mr{s,min} = \frac{2 f_\mr{H}}{k_\mr{max}}
$$

## <center> Izbor učestanosti odabiranja

$$
\frac{2}{k} \frac{f_\mr{H}}{B} 
\leq 
\frac{f_\mr{s}}{B} 
\leq
\frac{2}{k-1}\left( \frac{f_\mr{H}}{B} -1 \right)
$$

## <center> Primer odabiranja signala ograničenog u spektru

A/D konvertor se koristi za odabiranje signala

$$
y(t) = A_1 \sin(2 \pi f_1 t) + A_2 \sin(2 \pi f_2 t)
$$

gde su $A_1=1$, $A_2 = 0.1$, $f_1 = 89$ MHz, $f_2 = 91$ MHz.
Za A/D konvertor napona pune skale $V_\mr{FS} = 1~\mr{V}$ očekivane spektralne komponente  signala su
$$
A_1 = 10 \log_{10} \left[ \frac{1}{2} \left( \frac{1}{\sqrt{2}} \frac{A_1}{V_\mr{FS}} \right)^2 \right]=-6~\mr{dBFs}
$$
$$
A_2 = 10 \log_{10} \left[ \frac{1}{2} \left( \frac{1}{\sqrt{2}} \frac{A_2}{V_\mr{FS}} \right)^2 \right]=-26~\mr{dBFs}
$$


Usvojićemo da je širina spektra ulaznog signala 3 MHz i centralna učestanost 90 MHz.
$$
f_\mr{L} = 88.5~\mr{MHz}
$$
$$
f_\mr{H} = 91.5~\mr{MHz}
$$
Učestanost odabiranja u prvoj Nikvistovoj zoni je
$$
f_\mr{s} \geq 2 f_\mr{H} = 183~\mr{MHz}
$$

Najviša Nikvistova zona u kojoj je moguće izvršiti odabiranje je
$$
k_\mr{max} = \left\lfloor \frac{f_\mr{H}}{B'} \right\rfloor = 30
$$
Učestanost odabiranja za 30. Nikvistovu zonu je u opsegu
\begin{align*}
\frac{2f_\mr{H}}{k} &\leq f_\mr{s} \leq \frac{2 f_\mr{L}}{k-1} \\
6.1~\mr{MHz}&\leq f_\mr{s} \leq 6.103~\mr{MHz}
\end{align*}
Učestanost odabiranja se može izabrati u veoma uskom opsegu.
Izabraćemo $f_\mr{s}=6.1~\mr{MHz}$.


Pošto se signal odabira u 30. Nikvistovoj zoni, spektar signala je invertovan.

```python
import numpy as np
import matplotlib.pylab as plt
from scipy.signal import *
from scipy.signal.windows import flattop
from scipy.signal.windows import hann

N = 65536
def genSines(t):
    f1 = 89.0e6
    f2 = 91.0e6
    A1 = 1
    A2 = 0.1
    y = A1*np.sin(2*np.pi*f1*t) + A2*np.sin(2*np.pi*f2*t)
    y *= win
    return y

win = hann(N)  # Using Hann window will introduce scalloping loss
winGain = 20*np.log10(N/sum(win))
fs = 183e6
T = 1.0/fs
freqs = (fs*np.arange(0, N))/N - fs/2
t = np.arange(0, N)*T
y = genSines(t)
Y = 20*np.log10(np.absolute(np.fft.fft(y)))
Y_hann = np.fft.fftshift(Y) - 20*np.log10(N) + winGain
plt.plot(freqs/1e6, Y_hann)
plt.xlabel("f [MHz]")
plt.ylabel("|Y(jf)| [dB]")
plt.grid()
plt.ylim(-80, 0)
plt.title("Odabiranje u prvoj Nikvistovoj zoni")
plt.show()
print(find_peaks(Y_hann, height=-30)[1])

fs = 6.1e6
T = 1.0/fs
freqs = (fs*np.arange(0, N))/N - fs/2
t = np.arange(0, N)*T
y = genSines(t)
Y = 20*np.log10(np.absolute(np.fft.fft(y)))
Y_hann_30th = np.fft.fftshift(Y) - 20*np.log10(N) + winGain
plt.plot(freqs/1e6, Y_hann_30th)
plt.xlabel("f [MHz]")
plt.ylabel("|Y(jf)| [dB]")
plt.grid()
plt.ylim(-80, 0)
plt.title("Odabiranje u 30. Nikvistovoj zoni")
plt.show();
print(find_peaks(Y_hann_30th, height=-30)[1])
```

# <center> Kvantizacija signala

Kvantizacija signala se vrši podelom raspona ulaznog signala u $L$ intervala
$$
I_k = \left\lbrace x_k < x < x_{k+1}  \right\rbrace,~k=1,2,\cdots,L
$$
pragovima odlučivanja $x_1, x_2, \cdots, x_\mr{L+1}$, gde su izlazne vrednosti kvantizatora $\hat{x}_1, \cdots, \hat{x}_{L}$
$$
x_\mr{q}(n) \equiv Q[x(n)] = \hat{x}_k,\text{ ako }x(n)\in I_k
$$

<img src="images/quantization_levels.svg" />

U sistemima za digitalnu obradu signala se uglavnom koristi uniformni (linearni) kvantizator. U tom slučaju su izlazne vrednosti kvantizatora:
$$
\hat{x}_{k+1}-\hat{x}_{k} = \Delta
$$
dok su pragovi odlučivanja:
$$
x_{k+1}-x_{k} = \Delta
$$

$N$-bitni kvantizator ima $L=2^N$ izlaznih vrednosti.


Za napon pune skale $V_\mr{FS}$ korak kvantizacije je
$$
\Delta = \frac{V_\mr{FS}}{2^N}
$$

$$
-\frac{\Delta}{2} < e_\mr{q}(n) \leq \frac{\Delta}{2}
$$

<img src="images/quantization_steps_err.svg" />

<img src="images/quantization_model.svg" />


Model greške kvantizacije uvodi sledeće pretpostavke
 - Greška kvantizacije $e_\mr{q}(n)$ je uniformna u intervalu $(-\frac{\Delta}{2}, \frac{\Delta}{2})$
 - Greška kvantizacije je stacionarni beli šum - greške kvantizacije $e_\mr{q}(n)$ i $e_\mr{q}(m)$ nisu korelisane za $m \neq n$
 - Greška kvantizacije $e_\mr{q}(n)$ nije korelisana sa signalom $x(n)$
 - Signal $x(n)$ je stacionaran i $E\left[ x(n)  \right]=0$


Uvedene pretpostavke nisu uvek zadovoljene
 - Npr. kod signala male amplitude - nekoliko LSB - greška kvantizacije je korelisana sa signalom


## <center> Snaga kvantizacionog šuma

$$
P_\mr{n} =E\left[ e_\mr{q}^2 \right]= \int\limits_{-\Delta/2}^{\Delta/2} e^2 p(e) \dd e =
\frac{1}{\Delta} \int\limits_{-\Delta/2}^{\Delta/2} e^2  \dd e = 
\frac{\Delta^2}{12}
$$


Snaga kvantizacionog šuma \textbf{ne zavisi} od učestanosti odabiranja.
$$
P_\mr{n} = \frac{\Delta^2}{12}
$$


__Dvostrana__ spektralna gustina snage kvantizacionog šuma je 
$$
N_\mr{e} = \frac{P_\mr{n}}{f_\mr{s}} = \frac{\Delta^2}{12 f_\mr{s}}
$$

## <center> Odnos signal - kvantizacioni šum

Maksimalna amplituda signala je $A = \frac{V_\mr{FS}}{2}=\frac{\Delta 2^N}{2}$
$$
v(t) = \frac{\Delta 2^N}{2} \sin(2 \pi f_0 t)
$$
Efektivna vrednost signala je $A/\sqrt{2}$, odakle je snaga signala
$$
P_\mr{s} = \left( \frac{\Delta 2^N}{2\sqrt{2}} \right)^2 = \Delta^2 2^{2N-3}
$$
Odnos signal-kvantizacioni šum ($SQNR$)
\begin{align*}
SQNR &= 10 \log_{10} \frac{P_\mr{s}}{P_\mr{n}} 
= 10 \log_{10} \frac{\Delta^2 2^{2N-3}}{\Delta^2/12} \\
&= 10 \log_{10} \frac{12}{8} 2^{2N} 
= 6.02 N + 1.76
\end{align*}


Odnos signal-kvantizacioni šum ($SQNR$)
$$
SQNR = 6.02 N + 1.76~\text{[dB]}
$$
uzima u obzir samo kvantizacioni šum.


Povećavanjem rezolucije A/D konvertora za jedan bit SNR se povećava za 6 dB.



U realnim A/D konvertorima postoje i drugi izvori šuma: termički šum, šum prekidanja, fazni šum...


```python
import numpy as np

class quantizator():
    def __init__(self, nbits, offset = 0.5):
        self.nbits = nbits
        self.offset = offset
        self.levels = -2**(nbits-1)+1.0*np.arange(0,2**nbits+1,1)+offset
        self.levels[-1] = np.inf
        self.delta = 2.0/2**(nbits)
        self.missingCodes = np.zeros(2**nbits+1)

    def stepErrors(self, maxErr):
        self.levels += (np.random.uniform(-maxErr, maxErr, (1,2**self.nbits+1)))[0]

    def quantizeNormalized(self, data):
        level = np.argmax(self.levels > data*2**(self.nbits-1))
        level += self.missingCodes[level]
        level += -2**(self.nbits-1)
        return self.delta*level

    def quantizeInteger(self, data):
        level = np.argmax(self.levels > data*2**(self.nbits-1))
        level += self.missingCodes[level]        
        level += -2**(self.nbits-1)
        return level

    def __call__(self, data, normalize = True):
        if normalize:
            vquant = np.vectorize(self.quantizeNormalized)
        else:
            vquant = np.vectorize(self.quantizeInteger)
        return vquant(data)
```

```python
import matplotlib.pylab as plt
import numpy as np
nBits = 4
nPoints = 16
a=quantizator(4,offset=0.5)
x=np.arange(-nPoints*2**nBits,nPoints*2**nBits,1)*1.0/(nPoints*2**nBits)

#a.stepErrors(0.2)
#a.missingCodes[32] = 1

y=a(x)
plt.plot(x, y)
plt.plot(x, x)
plt.xlabel(r'$V_\mathrm{in}$')
plt.ylabel(r'$V_\mathrm{out}$')
plt.show()
plt.plot(x,(y-x)/(a.delta))
plt.xlabel(r'$V_\mathrm{in}$')
plt.ylabel(r'$\varepsilon/\Delta$')
plt.show();
```

## <center> Nesupstraktivni diter

```python
import numpy as np
import matplotlib.pylab as plt
Nbits = 8
Npoints = 4096
M = 127
dM = 15
T = 1
quant = quantizator(Nbits)
quant.stepErrors(0.05)
A = 0.5 * quant.delta
t = T*np.arange(0,Npoints,1)
y1 = A*np.sin(2*np.pi*t*M/Npoints)
y2 = A*np.sin(2*np.pi*t*(M+dM)/Npoints)
for Adither in [0, 0.5, 1, 2]:
    dither = Adither * quant.delta * np.random.randn(Npoints)
    y = quant(y1 + y2 + dither)
    plt.plot(y/quant.delta)
    #y = y-dither
    plt.plot(y/quant.delta)
    plt.show()
    Y = 20 * np.log10(1e-25+abs(np.fft.fft(y)))-20*np.log10(Npoints)
    plt.plot(Y[0:int(Npoints/2)])
    fftNoiseFloor = 6.02*Nbits+1.76+10*np.log10(Npoints/2)
    plt.plot([0,int(Npoints/2)],[-fftNoiseFloor, -fftNoiseFloor])
    plt.show()
```

## <center> Diter van opsega signala
<img src="images/adc_outofband_dither.svg" />

```python
import numpy as np
import matplotlib.pylab as plt
from scipy.signal import *
Nbits = 8
Npoints = 4096
M = 1511
dM = 15
T = 1
quant = quantizator(Nbits)
quant.stepErrors(0.15)
A = 1 * quant.delta
t = T*np.arange(0,Npoints,1)
y1 = A*np.sin(2*np.pi*t*M/Npoints)
y2 = A*np.sin(2*np.pi*t*(M+dM)/Npoints)
nord, beta = kaiserord(60, 0.2)
lpf = firwin(nord, 0.1, window=('kaiser', beta))
w, h = freqz(lpf, 1, 3140)
plt.plot(w, 20*np.log10(np.absolute(h)))
plt.title('Filtar propusnik niskih učestanosti')
plt.grid()
plt.show()
for Adither in [0, 1, 5, 10, 50]:
    dither = Adither*quant.delta*np.random.randn(Npoints)
    dither = lfilter(lpf, 1, dither)
    y = quant(y1 + y2 + dither)
    Y = 20 * np.log10(1e-25+abs(np.fft.fft(y)))-20*np.log10(Npoints)
    plt.plot(Y[0:int(Npoints/2)])
    fftNoiseFloor = 6.02*Nbits+1.76+10*np.log10(Npoints/2)
    plt.plot([0,int(Npoints/2)],[-fftNoiseFloor, -fftNoiseFloor])
    plt.ylim(-fftNoiseFloor-10, 0)
    plt.title('Adither = '+str(Adither))
    plt.grid()
    plt.show()
```

# <center> Prozorske funkcije
    
U slučajevima kada nije moguće koristiti koherentno odabiranje signala dolazi do curenja spektra, i potrebno je koristiti prozorske funkcije.

## <center> Ekvivalentni propusni opseg šuma

Ekvivalentni propusni opseg šuma je širina idealnog filtra  koji na svom izlazu ima istu snagu šuma kao i realni filtar.
$$
f_\mr{enbw} = \dfrac{\int_{0}^{\infty} |H(\jj f)|^2 \dd f}{|H(0)|^2}
$$
<img src="images/enbw_def.svg" />

Ekvivalentni propusni opseg šuma FIR filtra je
$$
F_\mr{enbw,FIR} = \dfrac{\int\limits_{-1/2}^{1/2} |H(\jj F)|^2 \dd F}{|H(0)|^2}
$$
Pošto su granice integracije $(-1/2,1/2)$ izračunava se __dvostrani__ ekvivalentni propusni opseg šuma.

Impulsni odziv FIR filtra $N$-og reda je
$$
H(z) = \sum\limits_{n=0}^{N} b(n) z^{-n}
$$
Frekvencijski odziv u ustaljenom stanju se dobija za $z=e^{\jj 2 \pi F}$
$$
H(\jj F) = \sum\limits_{n=0}^{N} b(n) e^{-\jj 2\pi n F }
$$
odakle je $|H(0)|^2$
$$
|H(0)|^2 = \left| \sum\limits_{n=0}^{N} b(n) \right|^2
$$

Primenom Parsevalove teoreme 
$$
\int\limits_{-1/2}^{1/2} |H(\jj F)|^2 \dd F = \sum\limits_{n=0}^{N} |b(n)|^2
$$
dobijamo izraz za ekvivalentni propusni opseg šuma FIR filtra
$$
F_\mr{enbw,FIR} = \dfrac
{\sum\limits_{n=0}^{N} |b(n)|^2}
{\left|\sum\limits_{n=0}^{N} b(n) \right|^2}
$$
$ENBW$ se može izraziti relativno u odnosu na širinu bina
$$
\dfrac{F_\mr{enbw,FIR}}{F_\mr{bin}} = N \dfrac
{\sum\limits_{n=0}^{N} |b(n)|^2}
{\left|\sum\limits_{n=0}^{N} b(n) \right|^2}
$$

## <center> Pojačanje prozorske funkcije
    
Spektar signala dobijen diskretnom Furijeovom transformacijom u $N$ tačaka je skaliran sa $N$. Pored toga, pojačanje prozorske funkcije je
$$
|H(0)| = \left|\sum\limits_{n=0}^{N} b(n)\right|
$$
odakle je ukupan faktor skaliranja DFT-a prozorske funkcije
$$
k = \frac{N}{\left|\sum\limits_{n=0}^{N} b(n)\right|}
$$

```python
import pandas
import numpy as np
import matplotlib.pylab as plt
from scipy.signal import *
from scipy.signal.windows import *

def enbw(data):
    return len(data)*np.sum(data**2)/np.sum(data)**2

def gain(data):
    return len(data)/np.sum(data)

windowFns = ['hamming', 'hann', 'triang', 'blackman', 'bartlett', ('kaiser', 4.0), 'flattop' ]
N = 65536
res = []
cols = ['ENBW', 'Gain', 'Gain [dB]']
nFreqPoints = 1000
nBins = 10
wfreqs = 2*np.pi*np.arange(0, nFreqPoints)/(N*nFreqPoints/nBins)
for win in windowFns:
    data = get_window(win, N)
    winENBW = round(enbw(data), 2)
    winGain = round(gain(data), 2)
    winGaindB = round(20*np.log10(gain(data)), 2)
    res.append((winENBW, winGain, winGaindB))
    w, H = freqz(data,1, wfreqs)
    H = 20*np.log10(np.absolute(H))
    H = H + winGaindB - 20*np.log10(N)
    plt.plot(wfreqs/(2*np.pi)*N, H, aa=True, linewidth=2.5)
plt.grid()
plt.ylim(-120,5)
plt.xlim(0, nBins)
plt.xlabel('F [bin]')
plt.ylabel('|H(j f)| [dB]')
plt.show()

resTable = pandas.DataFrame(res, columns=cols, index = windowFns)
print(resTable);
```

## <center> Neizvesnost amplitude spektralne komponente (Scalloping loss)

Diskretna Furijeova transformacija $N$ odbiraka rezultuje spektrom sa $N$ frekvencijskih binova, gde je širina frekvencijskog bina $f_\mr{bin} = f_\mr{s}/N$.
Pošto je spektar signala diskretan i podeljen u $N$ binova, nije moguće razlikovati signale čije učestanosti su u okviru jednog bina.
Frekvencijski odziv prozorske funkcije u okviru bina
dovodi do neizvesnosti amplitude spektralne komponente (scalloping loss).

Prozorske funkcije koje su projektovane za malu neizvesnost amplitude spektralne komponente se nazivaju flat top prozorske funkcije.
U poređenju sa drugim prozorskim funkcijama, flat top prozorske funkcije imaju širi ekvivalentni propusni opseg šuma.
Flat top prozorske funkcije se koriste u primenama gde je tačnost amplitude od primarnog interesa, npr. u analizatorima spektra.

```python
import pandas
import numpy as np
import matplotlib.pylab as plt
from scipy.signal import *
from scipy.signal.windows import *

def enbw(data):
    return len(data)*np.sum(data**2)/np.sum(data)**2

def gain(data):
    return len(data)/np.sum(data)

windowFns = ['hamming', 'hann', 'triang', 'blackman', 'bartlett', ('kaiser', 4.0), 'flattop' ]
N = 65536
nFreqPoints = 1000
nBins = 1
wfreqs = 2*np.pi*np.arange(0, nFreqPoints)/(N*nFreqPoints/nBins)
cols = ['ENBW', 'Gain', 'Gain [dB]', 'Scalloping loss']
res = []
for win in windowFns:
    data = get_window(win, N)
    w, H = freqz(data,1, wfreqs)
    H *= gain(data)/N
    Hdb = 20*np.log10(np.absolute(H))
    winENBW = round(enbw(data), 2)
    winGain = round(gain(data), 2)
    winGaindB = round(20*np.log10(gain(data)), 2)
    scallopingLoss = Hdb[int(len(w)/2)]
    res.append((winENBW, winGain, winGaindB, scallopingLoss))    
    plt.figure(1)
    plt.plot(wfreqs/(2*np.pi)*N, Hdb, aa=True, linewidth=2.5)
    plt.figure(2)
    H2 = np.absolute(H)
    H2db = 20*np.log10(H2)
    plt.plot(wfreqs/(2*np.pi)*N, H2db, aa=True, linewidth=2.5)
    H2 = np.absolute(np.flip(H))
    H2db = 20*np.log10(H2)
    plt.plot(wfreqs/(2*np.pi)*N, H2db, aa=True, linewidth=2.5)
plt.figure(1)
plt.grid()
plt.ylim(-5,1)
plt.xlim(0, nBins)
plt.xlabel('F [bin]')
plt.ylabel('|H(j f)| [dB]')
plt.figure(2)
plt.grid()
plt.ylim(-5,1)
plt.xlim(0, nBins)
plt.xlabel('F [bin]')
plt.ylabel('|H(j f)| [dB]')
plt.show()
resTable = pandas.DataFrame(res, columns=cols, index = windowFns)
print(resTable);
```

## <center> Degradacija $SNR$ usled džitera referentnog takta

Usled šuma u signalu takta odabiranja i praga odlučivanja dolazi do promene trenutka odabiranja
<img src="images/adc_aperture_noise.svg" />

Promena trenutka odabiranja usled šuma u signalu takta i praga odlučivanja se naziva podrhtavanje ili džiter (enlg. jitter).
Džiter takta je slučajan proces sa Gausovom raspodelom i usko je povezan sa faznim šumom takta odabiranja.

<img src="images/clk_ideal_real.svg" />


<img src="images/aperture_jitter.svg" />

$$
v(t) = A \sin 2\pi f t
$$

$$
\frac{\dd v(t)}{\dd t} = 2\pi f A \cos 2\pi f t
$$

$$
\frac{\dd v(t)}{\dd t}\Big|_{\mr{max}} = 2\pi f A
$$

Efektivna vrednost promene napona usled džitera aperture
\begin{align*}
\Delta V_\mr{eff} &= \frac{1}{\sqrt{2}}\frac{\dd v(t)}{\dd t}\Big|_{\mr{max}} t_\mr{j} \\
         &= \sqrt{2} \pi f A t_\mr{j}
\end{align*}
$\Delta V_\mr{eff}$ je efektivna vrednost šuma usled džitera aperture


Ulazni signal $v(t) = A \sin 2 \pi f t$ ima efektivnu vrednost $V_\mr{sin,eff}=A/\sqrt{2}$.
Odnos signal-šum usled džitera aperture
\begin{align*}
SNR &= 10 \log_{10} \dfrac{P_\mr{signal}}{P_\text{šum}} \\
    &= 20 \log_{10} \dfrac{V_\mr{sin,eff}}{\Delta V_\mr{eff}} \\
    &= 20 \log_{10} \dfrac{A/\sqrt{2}}{2 \pi f t_\mr{j} A  / \sqrt{2}} \\
    &= 20 \log_{10} \dfrac{1}{2 \pi f t_\mr{j}}
\end{align*}

Usled neizvesnosti trenutka odabiranja $SNR$ je konačan i opada sa porastom učestanosti.



<img src="images/adc_jitter_snr.svg" />

Brzi konvertori (>100 MHz) visoke rezolucije (>8b) imaju stroge zahteve za faznim šumom.


Ukupan $SNR$ usled kvantizacije signala, termičkog šuma i džitera je dat sa


$$
SNR = - 10 \log_{10} \left(
10^{-\frac{SNR_\mr{kvant}}{20}}
+10^{-\frac{SNR_\text{šum}}{20}}
+10^{-\frac{SNR_\text{džiter}}{20}}
\right)
$$


Na niskim učestanostima $SNR$ je ograničen šumom kvantizacije ili termičkim šumom, dok na visokim učestanostima ograničenje uglavnom potiče od džitera.



Idealni takt odabiranja 
$$
v(t) = A \sin (\omega t)
$$


Realni takt odabiranja
$$
v(t) = A \sin (\omega t + \phi(t))
$$

Spektar idealnog takta odabiranja ima samo jednu diskretnu komponentu na učestanosti 
$\omega$.


Realni takt odabiranja je fazno modulisan slučajnim varijacijama faze $\phi(t)$. 
Usled fazne modulacije, spektar realnog takta odabiranja sadrži i bočne opsege - fazni šum.


Pored slučajnih varijacija faze, $\phi(t)$ može sadržati i periodične signale koje rezultuju diskretnim spektralnim komponentama - spurovima.
Fazno sinhronisane petlje (PLL) generišu spurove na učestanostima koje su umnošci referentne učestanosti.




$SQNR$ A/D konvertora sa $N$ bita rezolucije je
$$
SQNR = 6.02 N + 1.76
$$

$SNR$ signala na učestanosti $f$ usled džitera referentnog takta je
$$
SNR = -20 \log_{10} {2 \pi f t_j} = -20 \log_{10} 2\pi t_j - 20 \log_{10} f
$$

Kada je $SNR = SQNR$ dolazi do degradacije od 3~dB - 0.5 bita


Ekvivalentni broj bita usled džitera referentnog takta je
$$
ENOB = - \frac{1.76 + 20 \log_{10} 2 \pi f t_j}{6.02}
$$

Maksimalna učestanost ulaznog signala za $N$ bita rezolucije je
$$
f = 10^{-\frac{6.02 N + 1.76 + 20 \log_{10} 2 \pi t_j}{20}}
$$


```python
import matplotlib.pylab as plt
import numpy as np
from scipy.signal import *
Npoints = 4096
M = 2000 # input signal frequency FFT bin
T = 1e-9 # sampling time
fs = 1.0/T # sampling frequency
freqs = fs*np.arange(0, Npoints/2)/Npoints/1e6
fsig = fs * M / Npoints  # input signal frequency
t = T*np.arange(0,Npoints,1)
for tj in [0, 1e-13, 1e-12, 1e-11]:
    jitter = tj * np.random.randn(Npoints)
    y = np.sin(2*np.pi*fs*(t+jitter)*M/Npoints)
    Y = 20 * np.log10(1e-25+abs(np.fft.fft(y)))-20*np.log10(Npoints)
    plt.plot(freqs, Y[0:int(Npoints/2)])
    if tj>0:
        SNR = 20 * np.log10(2*np.pi*tj*fsig)
        plt.plot([0,freqs[-1]], [SNR, SNR], 'k', label=r'SNR')
        FFTPSD = SNR - 10*np.log10(Npoints/2)
        plt.plot([0,freqs[-1]], [FFTPSD, FFTPSD], 'r', label=r'FFT PSD')
        plt.legend(loc='upper left')
    plt.xlabel(r'f [MHz]')
    plt.ylabel(r'|Y(jf)| dB')
    plt.text(200, -10, 'tj='+str(tj*1e12)+' ps')
    plt.show();
```

## <center> Aproksimacija faznog šuma linearnim segmentima

Fazni šum se može aproksimirati linearnim segmentima, gde je 
$\mathcal{L}(f)$ u dBc/Hz, i koristi se logaritamska $f$  osa.

<img src="images/phasenoise_piecewise_linear.svg" />

U opsegu učestanosti $[f_i,f_{i+1})$ fazni šum se može aproksimirati

$$
L_i(f) = a_i \left( \log_{10} f - \log_{10} f_i \right) + b_i
$$


$$
a_i = \dfrac{L(f_{i+1})-L(f_{i})}{\log_{10} f_{i+1}- \log_{10} f_i}
$$


$$
b_i = L(f_i)
$$


$$
L(f) = 10 \log_{10} \mathcal{L}(f)
$$

Džiter takta odabiranja je

$$
t_\mr{j} = \frac{1}{\omega_0} 
\sqrt{ 2 \int\limits_{\Delta f_1}^{\Delta f_2} \mathcal{L}(f) \dd f}
$$

Korišćenjem aproksimacije linearnim segmentima dobijamo

$$
t_\mr{j} = \frac{1}{\omega_0} 
\sqrt{ 2 \sum\limits_{i=1}^{N-1} 
\int\limits_{\Delta f_i}^{\Delta f_{i+1}} \mathcal{L}_i(f) \dd f}
$$

Doprinos $i$-tog segmenta je

$$
t_{\mr{j},i} = \frac{1}{\omega_0} 
\sqrt{ 2 
\int\limits_{\Delta f_i}^{\Delta f_{i+1}} \mathcal{L}_i(f) \dd f}
$$

Ukupan džiter se dobija sumiranjem svih doprinosa

$$
t_{\mr{j}} = \sqrt{\sum\limits_{i=1}^{N-1} t_{\mr{j},i}^2}
$$


Doprinos $i$-tog segmenta se dobija integracijom faznog šuma

\begin{align*}
\int\limits_{\Delta f_i}^{\Delta f_{i+1}} \mathcal{L}_i(f) \dd f
&=
\int\limits_{\Delta f_i}^{\Delta f_{i+1}}
10^\frac{L(f)}{10}
\dd f \\
&=
\int\limits_{\Delta f_i}^{\Delta f_{i+1}}
10^\frac{a_i (\log_{10} f - \log_{10} f_i) +b_i) }{10}
\dd f
\end{align*}


$$
10^\frac{a_i (\log_{10} f - \log_{10} f_i) +b_i) }{10}
=
f^\frac{a_i}{10}
f_i^\frac{-a_i}{10}
10^\frac{b_i }{10}
$$

$$
\int\limits_{\Delta f_i}^{\Delta f_{i+1}} \mathcal{L}_i(f) \dd f
=
f_i^\frac{-a_i}{10}
10^\frac{b_i }{10}
\int\limits_{\Delta f_i}^{\Delta f_{i+1}}
f^\frac{a_i}{10}
\dd f
$$

$$
\int\limits_{\Delta f_i}^{\Delta f_{i+1}}
f^\frac{a_i}{10}
\dd f
=
\begin{cases}
\log f_{i+1} - \log f_{i}, & a_i = -10\\
\frac{1}{1+\frac{a_i}{10}} \left( f_{i+1}^{1+\frac{a_i}{10}} - f_{i}^{1+\frac{a_i}{10}} \right) , & a_i \neq -10
\end{cases}

## <center> Korekcija $\sin(x)/x$ frekvencijskog odziva kola zadrške nultog reda

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import *
fmax = 0.35 # max. frequency of interest
n = 8 # number of points for approximation
f = fmax * np.arange(0, n+1, 1)/n
H_sinc = np.sin(np.pi*f)/(np.pi*f+1e-15)
H_sinc[0] = 1.0
f_goal = np.zeros(2*(n+1))
f_goal[0::2] = f
f_goal[1::2] = f+1e-3
f_goal *= 2	# In firls function Nyquist frequency is 1
goal = np.zeros(2*(n+1))
goal[0::2] = 1.0/H_sinc
goal[1::2] = 1.0/H_sinc
h_invsinc = firls(5, f_goal, goal)
w,H = freqz(tuple(h_invsinc),tuple([1]))
H_sinc = np.sin(w/2)/(w/2+1e-15)
H_sinc[0] = 1.0
plt.plot(w/(2*np.pi),np.absolute(1.0/H_sinc), 'k', aa=True, linewidth=2.5,label=r'$x/\sin(x)$')
plt.plot(w/(2*np.pi),np.absolute(H), 'gray', aa=True, linewidth=2.5,label=r'$|H(\mathrm{j} F)|$')
plt.legend(loc='upper left')
df = 0.005
for i in range(0,len(goal),2):
    plt.plot((f_goal[i]/2-df,f_goal[i+1]/2+df),[goal[i], goal[i+1]], 'r',aa=True, linewidth=2.5)
plt.xlim(0,0.5)
plt.ylim(1,1.4)
plt.grid()
plt.xlabel(r'$F$')
plt.ylabel(r'$|H(\mathrm{j} F)|$')
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import *
N = 65536  # number of data points
M = 8  # oversample factor
# Generate random numbers to get flat spectrum
phase = np.random.uniform(-np.pi,np.pi,N)
Y = np.ones(int(N),dtype=complex) * np.exp(1j*phase)
y = np.fft.ifft(Y)
plt.plot(np.real(y))
plt.plot(np.imag(y))
plt.show()
nrz = 1.0*np.ones((1, M))[0]/(1.0*M)
rz = 1.0*np.ones((1, M))[0]/(1.0*M)
rz[int(M/2):] = 0
rf = 1.0*np.ones((1, M))[0]/(1.0*M)
rf[int(M/2):] = -rf[int(M/2):]
rfrz = 1.0*np.ones((1, M))[0]/(1.0*M)
rfrz[int(M/2):] = 0
rfrz[int(M/4):int(M/2)] = -rfrz[int(M/4):int(M/2)]
yr_nrz = lfilter(nrz, 1.0, y)
yr_rz = lfilter(rz, 1.0, y)
yr_rf = lfilter(rf, 1.0, y)
yr_rfrz = lfilter(rfrz, 1.0, y)
x = 1.0*np.arange(0, int(N/2), 1)/N*M
Yr_nrz = 20*np.log10(np.absolute(np.fft.fft(yr_nrz)))
Yr_rz = 20*np.log10(np.absolute(np.fft.fft(yr_rz)))
Yr_rf = 20*np.log10(np.absolute(np.fft.fft(yr_rf)))
Yr_rfrz = 20*np.log10(np.absolute(np.fft.fft(yr_rfrz)))
Yr = [Yr_nrz, Yr_rz, Yr_rf, Yr_rfrz]
names = ['dac_reconstruction_nrz', 'dac_reconstruction_rz', 'dac_reconstruction_rf', 'dac_reconstruction_rfrz']
labels = [r'$\mathrm{NRZ}$', r'$\mathrm{RZ}$', r'$\mathrm{RF}$', r'$\mathrm{RFRZ}$']
#labels = ['NRZ', 'RZ', 'RF', 'RFRZ']
for i in range(0, len(Yr)):
    plt.plot(x,Yr[i][0:int(N/2)], aa=True, linewidth=1.5, color='k')
    plt.ylim(-40,0)
    plt.xlim(0, 0.5*M)
    plt.xlabel(r"$f/f_\mathrm{s}$")
    plt.ylabel(r"$|H(\mathrm{j} \omega)|~\mathrm{[dB]}$")
    plt.text(3.5,-7.5,labels[i],fontsize=36,ha='center')
    plt.grid()
    plt.tight_layout()
    plt.show()
```
