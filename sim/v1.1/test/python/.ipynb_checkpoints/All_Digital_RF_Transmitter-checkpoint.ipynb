```python
from signal_utils import *

def fftdB(x, win=False):
    if win:
        x = x*signal.windows.hann(len(x), False)
    xfftdB = 20*np.log10(np.abs(np.fft.fftshift(np.fft.fft(x))) + 1e-15)
    xfftdB -= np.max(xfftdB)
    freqs = np.arange(len(x)) / len(x) - 0.5
    return freqs, xfftdB
```

```python
# Taylor delay parameters

k3 = np.array([0.5, 0.0, -0.5])
k5 = np.array([-1, 8, 0, -8, +1]) / 12.0
k7 = np.array([1, -9, +45, 0, -45, +9, -1]) / 60.0
k9 = np.array([-1, 12, -39, +56, 0, -56, +39, -12, +1]) / 280.0
kernel = k5

# Serialization (LUT) parameters
LUT = [LUT1, LUT2, LUT3, LUT4, LUT5]
n_lut = len(LUT)
lut_width = np.array([len(l[0]) for l in LUT])

# Interpolation parameters
OSR = 8
AdB_interpolation = 80

print(f"OSR = {OSR}")
print(f"AdB_interpolation = {AdB_interpolation}\n")

print(f"n_lut = {n_lut}")
print(f"lut_width = {lut_width}")
```

```python
# Taylor 1st Order Delay
i_delta = 1/8 * 1/lut_width * 1/OSR
q_delta = -1/8 * 1/lut_width * 1/OSR

x_delay, xi_delay, xq_delay = [], [], []
cnt = 0
for id_, qd_ in zip(i_delta, q_delta):
    xid = frac_delay_1st(xi_test, id_, kernel)
    xqd = frac_delay_1st(xq_test, qd_, kernel)
    # xid /= np.max(xid)
    # xqd /= np.max(xqd)
    xd = xid + 1j*xqd
    # print(np.max(xid))
    
    xi_delay.append(xid)
    xq_delay.append(xqd)
    x_delay.append(xd)

for i, xd in enumerate(x_delay):
    freqs_, xfftdB_ = fftdB(xd, True)
    xmin = -0.5
    xmax = 0.5
    plt.plot(freqs_, xfftdB_, label=f"LUT {i+1}")
    plt.plot([xmin, xmax], [-AdB_interpolation, -AdB_interpolation], '--r')
    plt.legend()
    plt.xlim([xmin, xmax])
    plt.grid()
    plt.show();
```

# Interpolation

```python
deltaPass = 10**(-AdB_interpolation/20)
deltaStop = deltaPass
Fpass = Fmax/OSR
Fstop = (1-Fmax)/OSR

firCoeff = remezlp(Fpass, Fstop, deltaPass, deltaStop, even_n = False, nPoints=N, Nmax=N)
firCoeff /= np.sum(firCoeff)

x_osr8, xi_osr8, xq_osr8 = [], [], []
for xi, xq in zip(xi_delay, xq_delay):
    # xio = np.real(interpolate(xi, OSR, F, delta, show=False, prt=False))
    # xqo = np.real(interpolate(xq, OSR, F, delta, show=False, prt=False))
    xio = polyI(xi, firCoeff, OSR)
    xqo = polyI(xq, firCoeff, OSR)
    xio = np.array(xio) * 8
    xqo = np.array(xqo) * 8
    xo = xio + 1j*xqo

    xi_osr8.append(xio)
    xq_osr8.append(xqo)
    x_osr8.append(xo)

for i, xo in enumerate(x_osr8):
    freqs_, xfftdB_ = fftdB(xo, True)
    xmin = -0.5
    xmax = 0.5
    plt.plot(freqs_, xfftdB_, label=f"LUT {i+1}")
    plt.plot([xmin, xmax], [-AdB_interpolation, -AdB_interpolation], '--r')
    plt.legend()
    plt.xlim([xmin, xmax])
    plt.grid()
    plt.show();
```

# Delta-Sigma Modulation

```python
# Path to C++ executable
cpp_path = "../build/delta-sigma"
xi_in_path = "./data/serial/xi_in.txt"
xq_in_path = "./data/serial/xq_in.txt"
xi_out_path = "./data/serial/xi_out.txt"
xq_out_path = "./data/serial/xq_out.txt"
iir_path = "../data/input/deltaSigma_iirs.txt"

x_ds, xi_ds, xq_ds = [], [], []
cnt = 0
for xi, xq in zip(xi_osr8, xq_osr8):
    if cnt <= 2:
        quant_typ = "mid-tread"
    else:
        quant_typ = "mid-rise"

    fact = 1/4
    xi = np.array(xi)/fact
    xq = np.array(xq)/fact
    
    np.savetxt(xi_in_path, xi, fmt="%.18e")
    np.savetxt(xq_in_path, xq, fmt="%.18e")
    
    # Call the C++ program with input and output file paths as arguments
    subprocess.run([cpp_path, xi_in_path, xq_in_path, xi_out_path, xq_out_path, iir_path, quant_typ])
    
    xids = np.round(np.loadtxt(xi_out_path))
    xqds = np.round(np.loadtxt(xq_out_path))
    xds = xids + 1j*xqds

    xi_ds.append(xids)
    xq_ds.append(xqds)
    x_ds.append(xds)
    
    cnt += 1
```

```python
for i, xds in enumerate(x_ds):
    if i <= 2:
        quant_typ = "mid-tread"
    else:
        quant_typ = "mid-rise"
        
    freqs_, xfftdB_ = fftdB(xds, True)
    xmin = -0.5
    xmax = 0.5
    plt.plot(freqs_, xfftdB_, label=f"LUT {i+1}, {quant_typ}")
    plt.plot([xmin, xmax], [-AdB_interpolation, -AdB_interpolation], '--r')
    plt.legend()
    plt.xlim([xmin, xmax])
    plt.grid()
    plt.show();
```
